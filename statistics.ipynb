```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# ✅ Load the Polyvore dataset (train.json)
json_file = "/home/nogaschw/outfit-transformer-main/datasets/polyvore/disjoint/train.json"


with open(json_file, "r") as f:
    train_data = json.load(f)

# ✅ Extract number of items per outfit
outfits = [{"set_id": outfit["set_id"], "num_items": len(outfit["item_ids"])} for outfit in train_data]

# ✅ Convert to DataFrame
df_outfits = pd.DataFrame(outfits)

# ✅ Count total outfits
total_outfits = len(df_outfits)

# ✅ Count outfits with 4-7 items
valid_outfits = df_outfits[(df_outfits["num_items"] >= 4) & (df_outfits["num_items"] <= 7)]
num_valid_outfits = len(valid_outfits)

# ✅ Compute the percentage
percentage = (num_valid_outfits / total_outfits) * 100

# ✅ Print results
print(f"✅ Total outfits: {total_outfits}")
print(f"✅ Outfits with 4-7 items: {num_valid_outfits} ({percentage:.2f}%)")

# ✅ Generate a histogram of outfit sizes
plt.figure(figsize=(8, 5))
plt.hist(df_outfits["num_items"], bins=range(1, max(df_outfits["num_items"]) + 2), edgecolor="black", alpha=0.7)
plt.axvline(4, color="r", linestyle="dashed", linewidth=1, label="Min 4 items")
plt.axvline(7, color="r", linestyle="dashed", linewidth=1, label="Max 7 items")
plt.xlabel("Number of Items in Outfit")
plt.ylabel("Frequency")
plt.title("Distribution of Items per Outfit in Polyvore Dataset")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
